```python
import pandas as pd
import numpy as np
from data_explorer import DataExplorer
```

```python
np.random.seed(42)
n = 200

uid   = np.random.randint(1e6, 1e7, n)
names = np.random.choice(['Alice', 'Bob', 'Charlie', None], n)
sex   = np.random.choice(['Male', 'Female'], n)
vals  = np.random.randint(1, 100, n)
group = np.random.choice([1, 2, 3, np.nan], n)
dates = np.random.choice(pd.to_datetime(['2020-07-02', '2020-07-03', '2020-07-05', '2020-07-07']), n)
df = pd.DataFrame({"id": uid, "name": names, "sex": sex, "group": group, "value": vals, "date": dates})
df.head()
```

```python
de = DataExplorer(df)
de
```
